```python
import pandas as pd
import pickle

from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)

from sklearn.ensemble import RandomForestClassifier
from imblearn.combine import SMOTEENN
from xgboost import XGBClassifier
```

```python
df = pd.read_csv('/content/drive/MyDrive/Customer_churn_pred/tel_churn.csv')
```

```python
df.head(5)
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
df.head()
```

```python
X = df.drop('Churn', axis=1)
X
```

```python
Y = df['Churn']
Y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2,random_state=42,
    stratify=Y)
```

```python
#now random forest classifier

from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf = RandomForestClassifier(n_estimators=100, criterion='gini', random_state=100, max_depth=6,min_samples_leaf=8)
model_rf.fit(x_train, y_train)
y_pred_rf = model_rf.predict(x_test)
```

```python
y_prob_rf = model_rf.predict_proba(x_test)[:, 1]
```

```python
print(classification_report(y_test, y_pred_rf, labels=[0,1]))
```

this shows that the data is imbalanced so we used smoteen

```python
sm = SMOTEENN()
X_resampled, y_resampled = sm.fit_resample(x_train, y_train)
```

```python
xr_train, xr_test, yr_train, yr_test = train_test_split(X_resampled, y_resampled, test_size=0.2)
```

```python

model_smote_rf = RandomForestClassifier(n_estimators=100, criterion='gini',random_state=100,max_depth=6,min_samples_leaf=8)
```

```python
model_smote_rf.fit(xr_train, yr_train)
y_pred_smote_rf = model_smote_rf.predict(xr_test)
```

```python
y_prob_smote_rf = model_smote_rf.predict_proba(x_test)[:, 1]
```

```python
print(classification_report(yr_test, y_pred_smote_rf, labels=[0,1]))
```

this shows that this model s much much bette than the previous one

```python
print(confusion_matrix(yr_test, y_pred_smote_rf))
```

```python
# model_xgb = XGBClassifier(
#     n_estimators=300,
#     max_depth=5,
#     learning_rate=0.05,
#     subsample=0.8,
#     colsample_bytree=0.8,
#     random_state=42,
#     scale_pos_weight= (len(y_train) - sum(y_train)) / sum(y_train)
# )

model_xgb = XGBClassifier(
    n_estimators=200,
    max_depth=5,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    eval_metric='logloss'
)
```

```python
model_xgb.fit(xr_train, yr_train)
```

```python
y_pred_xgb = model_xgb.predict(xr_test)
y_prob_xgb = model_xgb.predict_proba(xr_test)[:, 1]
```

```python

print("===== XGBOOST (SMOTEENN) =====")
print("Accuracy:", accuracy_score(yr_test, y_pred_xgb))
print("Precision:", precision_score(yr_test, y_pred_xgb))
print("Recall:", recall_score(yr_test, y_pred_xgb))
print("F1-score:", f1_score(yr_test, y_pred_xgb))
print("ROC-AUC:", roc_auc_score(yr_test, y_prob_xgb))

print(classification_report(yr_test, y_pred_xgb, labels=[0,1]))
print(confusion_matrix(yr_test, y_pred_xgb))
```

```python
rf_f1 = f1_score(yr_test, y_pred_smote_rf)
xgb_f1 = f1_score(yr_test, y_pred_xgb)

print(rf_f1)
print(xgb_f1)
```

```python
print("\n===== MODEL COMPARISON =====")
print("Random Forest F1:", rf_f1)
print("XGBoost F1:", xgb_f1)

if xgb_f1 > rf_f1:
    best_model = model_xgb
    model_name = "XGBoost"
else:
    best_model = model_smote_rf
    model_name = "Random Forest"

print("Best Model:", model_name)
```

```python
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score

best_threshold = 0
best_f1 = 0
best_precision = 0
best_recall = 0

thresholds = np.arange(0.1, 0.9, 0.01)

for t in thresholds:
    y_pred = (y_prob > t).astype(int)

    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    # condition: keep recall at least 0.70
    if f1 > best_f1 and recall >= 0.70:
        best_f1 = f1
        best_threshold = t
        best_precision = precision
        best_recall = recall

print("Best Threshold:", best_threshold)
print("Precision:", best_precision)
print("Recall:", best_recall)
print("F1-score:", best_f1)
```
